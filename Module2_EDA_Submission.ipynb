### Variables Description

4.Reviews - It tells us about the total number of users who have given a review for the application.

5.Size - It tells us about the size being occupied the application on the mobile phone.

6.Installs - It tells us about the total number of installs/downloads for an application.

7.Type - IIt states whether an app is free to use or paid.

8.Price - It gives the price payable to install the app. For free type apps, the price is zero.

9.Content Rating - It states whether or not an app is suitable for all age groups or not.

10.Genres - It tells us about the various other categories to which an application can belong.

11.Last Updated - It tells us about the when the application was updated.

12.Current Ver - It tells us about the current version of the application.

13.Android Ver - It tells us about the android version which can support the application on its platform

B.) User review Dataset

Let us first define what information the columns contain based on our inspection.

user_reviews dataframe has 64295 rows and 5 columns. The 5 columns are identified as follows:

1.App: Contains the name of the app with a short description (optional).

2.Translated_Review: It contains the English translation of the review dropped by the user of the app.

3.Sentiment: It gives the attitude/emotion of the writer. It can be ‘Positive’, ‘Negative’, or ‘Neutral’.

4.Sentiment_Polarity: It gives the polarity of the review. Its range is [-1,1], where 1 means ‘Positive statement’ and -1 means a ‘Negative statement’.

5.Sentiment_Subjectivity: This value gives how close a reviewers opinion is to the opinion of the general public. Its range is [0,1]. Higher the subjectivity, closer is the reviewers opinion to the opinion of the general public, and lower subjectivity indicates the review is more of a factual information.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# from the table above we can conclude that there are 1474 null values in the column called Rating now we are going to fill all the null spaces with the mean of the respective column
# .fillna method is used to fill null spaces with repective values that we intend to
playstore_data['Rating'] = playstore_data['Rating'].fillna(value = 4.193)
````

````python
print(playstore_data['Rating'].isnull().sum())
````

````python
# now we need to deal with the null values in other columns
# dealing with column = 'Current Ver' , 'Android Ver' , 'Type' , 'Content Rating'
playstore_data['Current Ver'] = playstore_data['Current Ver'].fillna(value = 'ROBLOX')
playstore_data['Android Ver'] = playstore_data['Android Ver'].fillna(value = '4.1 and up')
playstore_data['Type'] = playstore_data['Type'].fillna(value = 'Free')
playstore_data['Content Rating'] = playstore_data['Content Rating'].fillna(value = 'Everyone')
````

````python
# Checking weather all the null vakues have been eliminated
print(playstore_data['Current Ver'].isnull().sum())
print(playstore_data['Android Ver'].isnull().sum())
print(playstore_data['Type'].isnull().sum())
print(playstore_data['Content Rating'].isnull().sum())
````

````python
## converting the datatype of column 'Install' (from object to float)
playstore_data['Installs']= playstore_data['Installs'].apply((lambda x : str(x).replace('+', " ") if '+' in str(x) else str(x)))
playstore_data['Installs']= playstore_data['Installs'].apply((lambda x : str(x).replace(',', ''  ) if ',' in str(x) else str(x)))

# Removing the row where 'Installs' is 'Free' before converting to float
playstore_data = playstore_data[playstore_data['Installs'] != 'Free']
playstore_data['Installs']= playstore_data['Installs'].apply((lambda x : float(x)))
````

````python
## Preparing Dataframe which contains 'Catergory','Installs' counts
category_installs_sum_df= playstore_data.groupby(['Category'])[['Installs']].sum().reset_index()
category_installs_sum_df
````

````python
# cleaning the 'Size' column and convert the datatype "object" to "float"
def correct_size(s):

  try:
    if s[-1]=='M':
      return float(s[:-1])
    elif s[-1]=='k':
      return  float(s[:-1])/1024 # 1MB= 1024KB, converting KB TO MB.
    elif s== '1000+':
      return 1.0
    else:
      return 0.0
  except:
    return  0.0
````

````python
playstore_data['Size']= playstore_data['Size'].apply(correct_size)
````

````python
Category_Size_sum_df= playstore_data.groupby(['Category'])[['Size']].median().reset_index()
Category_Size_sum_df
````

````python
## Converting the column 'Price' datatype from object to float
playstore_data['Price']= playstore_data['Price'].apply((lambda x : str(x).replace('$', " ") if '$' in str(x) else str(x)))
playstore_data['Price']= playstore_data['Price'].apply((lambda x : float(x)))
````

````python
## Preparing Datarframe which contains 'Free App, Installs' count
free_app_installs_df = playstore_data[playstore_data['Price']==0].groupby(['Genres'])[['Installs']].sum().reset_index()
free_app_installs_df
````

````python
## Preparing Datarframe which contains 'Paid App, Installs' count
paid_app_installs_df = playstore_data[playstore_data['Price']!=0].groupby(['Genres'])[['Installs']].sum().reset_index().rename(columns={'Installs' : 'Paid_Installs'})
paid_app_installs_df
````
